### EDA

```python
# Calculate counts for each booking status
booking_counts = df['is_canceled'].value_counts()

# Create a bar plot
plt.figure(figsize=(6, 4))
colors = ['red', 'blue']  # Colors for 'Cancelled' (1) and 'Confirmed' (0)
bars = plt.bar(booking_counts.index, booking_counts.values, color=colors, alpha=0.8)

# Set plot labels and title
plt.xlabel('Booking Type')
plt.ylabel('Number of Bookings')
plt.title('Number of Bookings: (Cancelled-1 & Confirmed-0)')

# Add data labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, str(yval), va='bottom' if yval < 0 else 'top', ha='center', fontsize=10, color='black')

# Customize x-axis labels
plt.xticks(booking_counts.index, labels=['Cancelled', 'Confirmed'])

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Calculate counts for each hotel type
hotel_counts = df['hotel'].value_counts()

# Create a bar plot
plt.figure(figsize=(6, 4))
colors = ['blue', 'orange']  # Colors for 'Resort' and 'City' hotels
bars = plt.bar(hotel_counts.index, hotel_counts.values, color=colors, alpha=0.8)

# Set plot labels and title
plt.xlabel('Hotel Type')
plt.ylabel('Number of Bookings')
plt.title('Number of Bookings by Hotel Type')

# Add data labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, str(yval), va='bottom' if yval < 0 else 'top', ha='center', fontsize=10, color='black')

# Customize x-axis ticks and labels
plt.xticks(range(len(hotel_counts.index)), ['Resort Hotel', 'City Hotel'])

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Group the data by 'hotel' and 'is_canceled' and count the occurrences
booking_counts = df.groupby(['hotel', 'is_canceled']).size().unstack(fill_value=0)

# Create a bar plot
plt.figure(figsize=(9, 4))
colors = ['blue', 'orange']  # Color for 'Cancelled' (1) and 'Confirmed' (0)
ax = booking_counts.plot(kind='bar', stacked=True, color=colors, alpha=0.8)

# Set plot labels and title
plt.xlabel('Hotel Type')
plt.ylabel('Number of Bookings')
plt.title('Resort vs City Hotel')

# Set x-axis ticks and labels
plt.xticks(rotation=0)
plt.legend(title='Booking Status', labels=['Confirmed', 'Cancelled'])  # Adding legend

# Add data labels on top of the bars
for patch in ax.patches:
    x = patch.get_x() + patch.get_width() / 2
    y = patch.get_y() + patch.get_height()
    value = int(patch.get_height())
    ax.annotate(value, (x, y), textcoords="offset points", xytext=(0,5), ha='center', fontsize=10, color='black')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
# Initialize LabelEncoder
encoder = LabelEncoder()

# Apply label encoding to object-type columns
object_columns = df.select_dtypes(include=['object']).columns
for column in object_columns:
    df[column] = encoder.fit_transform(df[column])

# Display the encoded DataFrame
df.head()
```

### Spliting the data

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns='is_canceled', axis = 1)
y = df['is_canceled']
print("Shape of X: ",X.shape)   
print("Shape of y: ",y.shape)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)


print("Shapes of X_train and X_test: ",X_train.shape, X_test.shape)
print("Shape of y_train and y_test: ",y_train.shape, y_test.shape)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
lr = LogisticRegression()
lr.fit(X_train, y_train)

X_train_predict = lr.predict(X_train)
y_pred = lr.predict(X_test)
training_data_accuracy = accuracy_score(X_train_predict, y_train)
print("Train accuracy is : ", training_data_accuracy)
```

```python
y_pred = lr.predict(X_test)
y=y_test
test_data_accuracy = accuracy_score(y,y_pred)
print("Test accuracy is : ", test_data_accuracy)
```

```python
cm = confusion_matrix(y,y_pred)
sns.heatmap(cm, annot=True)
print("\n Classification report\n")
print(classification_report(y,y_pred))
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dt_clf = dtc.fit(X_train,y_train)
X_train_predict = dtc.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predict, y_train)
print("Train accuracy is : ", training_data_accuracy)
y_pred = dtc.predict(X_test)
y=y_test
test_data_accuracy = accuracy_score(y,y_pred)
print("Test accuracy is : ", test_data_accuracy)
```

```python
cm1 = confusion_matrix(y,y_pred)
sns.heatmap(cm1, annot=True)
print("\n Classification report\n")
print(classification_report(y,y_pred))
```
